```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x * x
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titulo')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Titulo')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo')

ax[1].plot(y, x, 'g')
ax[1].set_title('Titulo')
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12, 3))
axes.plot(x, y, 'r-', label='x ^ 2')
axes.plot(y, x, 'g-', label='x ** 1/2')
axes.set_title('Titulo')
axes.legend(loc=0)
```
