# Plots matriciais

```python
import seaborn as sns
%matplotlib inline
```

```python
fights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
fights.head()
```

```python
tips.head()
```

```python
crr = tips.corr()
```

```python
sns.heatmap(crr, cmap='coolwarm', annot=True)
```

```python
fights.info()
```

```python
fig = fights.pivot_table(values='passengers', index='month', columns='year')
fig
```

```python
sns.heatmap(fig, cmap='magma', linecolor='white', linewidths=1)
```
