# Estilos e Cores

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True)
```

```python
sns.set_context('poster', font_scale=0.6)
sns.lmplot(x='total_bill', y='tip', data=tips, size=5, aspect=4, hue='sex', palette='inferno')
```
